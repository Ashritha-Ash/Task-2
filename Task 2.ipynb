```python
# SECTION 1: Variables & Data Types

# A variable stores a value. The type depends on the value assigned.

name = "Ashritha"          # String
age = 20                   # Integer
height = 5.5               # Float
is_student = True          # Boolean value

print("Name:", name)
print("Age:", age)
print("Height:", height)
print("Is Student:", is_student)
```

```python
# SECTION 2: Conditional Statements

number = 12   # Change to test with other numbers

if number > 0:
    print(number, "is Positive")
elif number < 0:
    print(number, "is Negative")
else:
    print("The number is Zero")
```

```python
# SECTION 3: Loops (for and while)

# FOR LOOP — prints numbers from 1 to 5
for i in range(1, 6):
    print("For Loop Value:", i)

print()  # Line break

# WHILE LOOP — prints numbers from 1 to 5
x = 1
while x <= 5:
    print("While Loop Value:", x)
    x += 1
```

```python
# SECTION 4: Functions

def greet(name):
    """This function greets the user with their name."""
    return f"Hello, {name}! Welcome to Python learning."

print(greet("Ashritha"))
```

```python
# EXERCISE 1: Swapping two variables

a = 5
b = 10

print("Before Swap:", a, b)

# Pythonic swap
a, b = b, a

print("After Swap:", a, b)
```

```python
# EXERCISE 2: Check even or odd

num = 15

if num % 2 == 0:
    print(num, "is Even")
else:
    print(num, "is Odd")
```

```python
# EXERCISE 3: Sum of first N numbers

n = 5
total = 0

for i in range(1, n + 1):
    total += i

print("Sum of first", n, "numbers is:", total)
```

```python
# EXERCISE 4: Simple Calculator using functions

def calculator(a, b, operator):
    """Performs basic arithmetic operations."""
    if operator == "+":
        return a + b
    elif operator == "-":
        return a - b
    elif operator == "*":
        return a * b
    elif operator == "/":
        return a / b
    else:
        return "Invalid operator"

print("10 + 4 =", calculator(10, 4, "+"))
```

```python
# EXERCISE 5: Print star pattern

rows = 5
for i in range(1, rows + 1):
    print("*" * i)
```

```python
# EXERCISE 6: Largest among three numbers

x, y, z = 45, 12, 78

largest = max(x, y, z)

print("The largest number is:", largest)
```

```python
# EXERCISE 7: Count digits in a number

num = 987654
count = 0

temp = num
while temp > 0:
    count += 1
    temp //= 10

print("Digits in", num, "=", count)
```

```python
# EXERCISE 8: Palindrome check

text = "level"

if text == text[::-1]:
    print(text, "is a Palindrome")
else:
    print(text, "is NOT a Palindrome")
```
